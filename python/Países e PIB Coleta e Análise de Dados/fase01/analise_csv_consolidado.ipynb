```python
# Leitura e análise do CSV consolidado

import pandas as pd
import matplotlib.pyplot as plt

# Carregar o arquivo CSV consolidado
df = pd.read_csv('/Coleta de Dados/fase01/dados_consolidados.csv')

# Exibe as primeiras linhas do DataFrame
print(df.head())

# Análise e estatisticas
pop_stats = df['populacao'].describe()
print("\nEstatísticas descritivas da população:")
print(pop_stats)

# MatplotLib
country_code = 'BRA'
country_name = df[df['codigo'] == country_code]['nome'].values[0]
years = [str(year) for year in range(1960, 2022)]
pib_values = df[df['codigo'] == country_code][years].values.flatten()

plt.figure(figsize=(10, 6))
plt.plot(years, pib_values, marker='o')
plt.title(f'Evolução do PIB de {country_name} (1960-2021)')
plt.xlabel('Ano')
plt.ylabel('PIB (em dólares)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
