```python
# Programa para calcular a duração de um jogo

# Lendo o horário de início
hora_inicio = int(input("Digite a hora de início (0-23): "))
minuto_inicio = int(input("Digite o minuto de início (0-59): "))

# Lendo o horário de fim
hora_fim = int(input("Digite a hora de fim (0-23): "))
minuto_fim = int(input("Digite o minuto de fim (0-59): "))

# Convertendo tudo para minutos
inicio_total_minutos = hora_inicio * 60 + minuto_inicio
fim_total_minutos = hora_fim * 60 + minuto_fim

# Calculando a duração
if fim_total_minutos < inicio_total_minutos:
    # Se o jogo termina no dia seguinte
    duracao_minutos = (24 * 60 - inicio_total_minutos) + fim_total_minutos
else:
    duracao_minutos = fim_total_minutos - inicio_total_minutos

# Convertendo de volta para horas e minutos
duracao_horas = duracao_minutos // 60
duracao_minutos = duracao_minutos % 60

print(f"\nDuração do jogo: {duracao_horas} horas e {duracao_minutos} minutos")
```

```python
# Programa para converter número em dia da semana

# Lendo o número do dia da semana
dia = int(input("Digite um número entre 1 e 7: "))

# Verificando se o número está no intervalo válido
if dia < 1 or dia > 7:
    print("\nErro: O número deve estar entre 1 e 7!")
else:
    # Convertendo o número em dia da semana
    if dia == 1:
        nome_dia = "Domingo"
    elif dia == 2:
        nome_dia = "Segunda-feira"
    elif dia == 3:
        nome_dia = "Terça-feira"
    elif dia == 4:
        nome_dia = "Quarta-feira"
    elif dia == 5:
        nome_dia = "Quinta-feira"
    elif dia == 6:
        nome_dia = "Sexta-feira"
    else:
        nome_dia = "Sábado"
    
    print(f"\nO dia da semana é: {nome_dia}")
```

```python

# Entrada das notas das 3 provas
nota1 = float(input("Nota da primeira prova: "))
nota2 = float(input("Nota da segunda prova: "))
nota3 = float(input("Nota da terceira prova: "))

# Entrada da frequência
frequencia = int(input("Frequencia: "))

# Cálculo do G1 (média aritmética das 3 provas)
g1 = (nota1 + nota2 + nota3) / 3

# Mostrando os resultados
print("\n=== Resultados ===")
print(f"G1: {g1:.2f}")
print(f"Frequência: {frequencia:.2f}%")

# Verificando a situação do aluno
if frequencia < 75:
    print("\nAluno REPROVADO por faltas!")
else:
    if g1 >= 7:
        print("\nAluno APROVADO em G1!")
    elif g1 < 4:
        print("\nAluno REPROVADO em G1!")
    else:
        # Aluno precisa fazer G2
        print("\nAluno precisa fazer exame (G2)")
        g2 = float(input("Digite a nota do exame (G2): "))
        
        # Calculando a média final
        media_final = (g1 + g2) / 2
        print(f"\nMédia final: {media_final:.2f}")
        
        if media_final >= 5:
            print("Aluno APROVADO em G2!")
        else:
            print("Aluno REPROVADO em G2!")
```

```python
import random

# Lista de cursos possíveis
cursos = ["Engenharia Civil", "Engenharia Elétrica", "Engenharia Mecânica", 
          "Engenharia de Computação", "Engenharia Química"]

# Lista para armazenar os dados dos estudantes
estudantes = []

# Gerando dados aleatórios para 50 estudantes
for estudante in range(50):
    idade = random.randint(18, 60)  # Idade entre 18 e 35 anos
    semestre = random.randint(1, 10)  # Semestre entre 1 e 10
    curso = random.choice(cursos)  # Curso aleatório da lista
    
    estudantes.append({
        "idade": idade,
        "semestre": semestre,
        "curso": curso
    })

# Calculando a média de idade
soma_idades = sum(estudante["idade"] for estudante in estudantes)
media_idade = soma_idades / len(estudantes)

# Encontrando o curso do aluno mais velho
aluno_mais_velho = max(estudantes, key=lambda x: x["idade"])
curso_mais_velho = aluno_mais_velho["curso"]

# Contando alunos no 5º semestre
alunos_quinto_semestre = sum(1 for estudante in estudantes if estudante["semestre"] == 5)

# Exibindo os resultados
print("\n=== Análise dos Estudantes ===")
print(f"Média de idade dos estudantes: {media_idade:.2f} anos")
print(f"Curso do aluno mais velho: {curso_mais_velho}")
print(f"Quantidade de alunos no 5º semestre: {alunos_quinto_semestre}")

# Exibindo todos os dados para verificação
print("\nDados completos dos estudantes:")
for i, estudante in enumerate(estudantes, 1):
    print(f"Estudante {i}:")
    print(f"  Idade: {estudante['idade']}")
    print(f"  Semestre: {estudante['semestre']}")
    print(f"  Curso: {estudante['curso']}")
    print()
```
