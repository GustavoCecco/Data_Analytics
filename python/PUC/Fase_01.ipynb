```python
print("=== Cadastro de Temperaturas Máximas de 2021 ===")

soma_temperaturas = 0
meses_escaldantes = 0

maior_temperatura = -61 
menor_temperatura = 51   

mes_mais_quente = 0
mes_menos_quente = 0

contador_mes = 1

while contador_mes <= 12:
    print("\nMês", contador_mes)

    # Validação da entrada da temperatura máxima
    while True:
        try:
            temp = float(input("Digite a temperatura máxima do mês (entre -60 e 50 ºC): "))
            if temp < -60 or temp > 50:
                print("Temperatura inválida! Deve estar entre -60 e 50 graus Celsius. Digite novamente.")
                continue
            break
        except:
            print("Entrada inválida! Digite um número válido.")

    # Atualiza a soma para cálculo de média
    soma_temperaturas += temp

    # Contar meses escaldantes (> 33°C)
    if temp > 33:
        meses_escaldantes += 1

    # Verificar maior temperatura
    if temp > maior_temperatura:
        maior_temperatura = temp
        mes_mais_quente = contador_mes

    # Verificar menor temperatura
    if temp < menor_temperatura:
        menor_temperatura = temp
        mes_menos_quente = contador_mes

    contador_mes += 1


print("\n=== Relatório Anual ===")

# Calculo da média de temperatura
media = soma_temperaturas / 12
print("Temperatura média máxima anual: %.2f ºC" % media)
print("Quantidade de meses escaldantes (> 33 ºC):", meses_escaldantes)

# Nome dos meses por extenso, usando apenas condições:
print("Mês mais escaldante do ano:", end=" ")
if mes_mais_quente == 1:
    print("janeiro")
elif mes_mais_quente == 2:
    print("fevereiro")
elif mes_mais_quente == 3:
    print("março")
elif mes_mais_quente == 4:
    print("abril")
elif mes_mais_quente == 5:
    print("maio")
elif mes_mais_quente == 6:
    print("junho")
elif mes_mais_quente == 7:
    print("julho")
elif mes_mais_quente == 8:
    print("agosto")
elif mes_mais_quente == 9:
    print("setembro")
elif mes_mais_quente == 10:
    print("outubro")
elif mes_mais_quente == 11:
    print("novembro")
elif mes_mais_quente == 12:
    print("dezembro")

print("Mês menos quente do ano:", end=" ")
if mes_menos_quente == 1:
    print("janeiro")
elif mes_menos_quente == 2:
    print("fevereiro")
elif mes_menos_quente == 3:
    print("março")
elif mes_menos_quente == 4:
    print("abril")
elif mes_menos_quente == 5:
    print("maio")
elif mes_menos_quente == 6:
    print("junho")
elif mes_menos_quente == 7:
    print("julho")
elif mes_menos_quente == 8:
    print("agosto")
elif mes_menos_quente == 9:
    print("setembro")
elif mes_menos_quente == 10:
    print("outubro")
elif mes_menos_quente == 11:
    print("novembro")
elif mes_menos_quente == 12:
    print("dezembro")
```
